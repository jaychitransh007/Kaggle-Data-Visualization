```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon.head()
```

```python
pokemon.columns
```

## Univariate Plotting

#### Pokemon Type

```python
pokemon.type1.value_counts().plot(kind = 'bar');
```

```python
pokemon.type2.value_counts().plot(kind = 'bar');
```

#### HP Stat

```python
pokemon.hp.value_counts().sort_index().plot(kind = 'line');
```

#### Weight

```python
pokemon.weight_kg.plot(kind = 'hist');
```

## Bivariate Plotting

```python
pokemon.plot(x = 'attack', y = 'defense', kind= 'scatter');
```

```python
pokemon.plot(x = 'attack', y = 'defense', kind= 'hexbin', gridsize= 20);
```

```python
pokemon_stats_legendary = pokemon.groupby(['is_legendary','generation']).mean()[['attack', 'defense']]
pokemon_stats_legendary.plot(kind= 'bar', stacked = True);
```

```python
pokemon_stats_by_generation = pokemon.groupby('generation').mean()[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']]
pokemon_stats_by_generation.head()
pokemon_stats_by_generation.plot(kind= 'line')
```

### Plotting using Seaborn

##### Countplot

```python
sns.countplot(pokemon.type1);
```

```python
sns.countplot(pokemon.type2);
```

##### KDE Plot

```python
sns.kdeplot(pokemon.hp, cut = 3)
```

Compare this with the plot pandas plot

##### Distplot

```python
sns.distplot(pokemon.weight_kg.dropna(), bins = 50);
```

##### Jointplot

```python
sns.jointplot(x = 'attack', y = 'defense', kind= 'hexbin', gridsize = 25, data = pokemon)
```

##### Boxplot

```python
sns.boxplot(x = 'generation', y = 'hp', data = pokemon);
```

```python
sns.boxplot(x = 'generation', y = 'attack', data = pokemon);
```

```python
sns.boxplot(x = 'generation', y = 'defense', data = pokemon);
```

```python
sns.boxplot(x = 'generation', y = 'speed', data = pokemon);
```

```python
sns.violinplot(x = 'generation', y = 'attack', data = pokemon);
```

```python
sns.violinplot(x = 'generation', y = 'defense', data = pokemon);
```
